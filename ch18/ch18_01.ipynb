```python
import numpy as np
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
X = iris.data
Y = iris.target
```

```python
np.einsum('ij->j', X)
```

```python
np.einsum('ij->i', X)
```

```python
np.einsum('ij->', X)
```

```python
np.einsum('ij->ji', X)
```

```python
X3D = np.stack([X[Y==0],
               X[Y==1],
               X[Y==2]], axis=0)
```

```python
X3D_T = np.einsum('ijk->ikj', X3D)
```

```python
X3D_T
```

```python
np.einsum('ij,kj->ik', X, X)
```

```python
G = np.einsum('ij,ik->jk', X, X)
```

```python
G
```

```python
G_3D = np.einsum('ijk,ijm->ikm', X3D, X3D)
```

```python
G_3D
```

```python
n = X.shape[0]
mean_X = np.einsum('ij->j', X) / n
```

```python
X_c = X - mean_X
variance = np.einsum('ij,ij->j',X_c, X_c) / (n-1)
```

```python
cov_matrix = np.einsum('ij,ik->jk',X_c, X_c) / (n-1)
```
