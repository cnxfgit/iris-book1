```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num = 2000
```

```python
mu = np.array([0, 0])
rho = 0
Sigma = np.array([[1, rho],
                 [rho, 1]])
```

```python
X_binormal = np.random.multivariate_normal(mu, Sigma, size=num)
```

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(X_binormal[:,0],
          X_binormal[:,1],
          s=100,
          marker='.',
          alpha=0.5,
          edgecolors='w')
ax.set_aspect('equal', adjustable='box')
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)
ax.set_xticks((-3, 0, 3))
ax.set_yticks((-3, 0, 3))
```
