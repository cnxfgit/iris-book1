```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num = 2000
X_uniform = np.random.uniform(low=-3, high=3, size=(num, 2))

fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(X_uniform[:,0],
          X_uniform[:,1],
          s = 100,
          marker='.',
          alpha=0.5,
          edgecolors='w')

ax.set_aspect('equal', adjustable='box')
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)
ax.set_xticks((-3, 0, 3))
ax.set_yticks((-3, 0, 3))
```
