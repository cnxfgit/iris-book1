```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
X, y = load_iris(return_X_y=True)
```

```python
X_trans, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2
)
```

```python
def visualize(df):
    fig, axs = plt.subplots(1,2,
                            gridspec_kw={'width_ratios':[4,1]})
    sns.heatmap(df.iloc[:,0:-1],
               cmap='RdYlBu_r', yticklabels=False,
               cbar=False, ax=axs[0])
    sns.heatmap(df.iloc[:, [-1]],
               cmap='Set3', yticklabels=False,
               cbar=False, ax=axs[1])
```

```python
columns = ['Sepal length, X1', 'Sepal width, X2',
          'Petal length, X3', 'Petal width, X4',
          'Species']
```

```python
df_full = pd.DataFrame(np.c_[X, y],columns=columns)
visualize(df_full)
```

```python
df_train = pd.DataFrame(np.c_[X_trans,y_train],
                       columns=columns)
visualize(df_train)
```

```python
df_test = pd.DataFrame(np.c_[X_test, y_test],
                      columns=columns)
```

```python
visualize(df_test)
```
