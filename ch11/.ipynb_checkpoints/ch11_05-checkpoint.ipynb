```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris_sns = sns.load_dataset('iris')
```

```python
fig, ax = plt.subplots()
sns.heatmap(data=iris_sns.iloc[:, 0:-1],
           vmin=0, vmax=8,
           ax=ax,
           yticklabels=False,
           xticklabels=['Sepal length', 'Sepal width',
                       'Petal length', 'Petal wodth'],
           cmap='RdYlBu_r')
```
